```python
import pandas as pd
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import sweetviz as sv
import time
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
## importação das tabelas
df_processos = pd.read_csv('../Tabelas/Processos.CSV', sep = ';', encoding='utf-8') 
tabelaIBGE=pd.read_csv('../Tabelas/Tabela IBGE.csv', sep= ';', encoding='utf-8')
```

```python
df_processos['Data de entrada']
```

```python
df_processos.drop(columns='Unnamed: 0',inplace=True)
df_processos['Data de entrada']= df_processos['Data de entrada'].apply\
          (pd.to_datetime, format='%Y-%m-%d')
df_processos['Data adesao']= df_processos['Data adesao'].apply\
          (pd.to_datetime, format='%d/%m/%Y')
df_processos[['Data adesao']].fillna('')
df_processos['Processo']=df_processos['Processo'].astype(str)
```

```python
#juntar a tabela IBGE com df_processos
## colocando os dados do IBGE junto
df_processos=df_processos.merge(tabelaIBGE, how='left', on='UF')
```

```python
# criar uma coluna com os dias de contencioso
df_processos['dias da situacao'] = datetime.strptime('29-10-2021', '%d-%m-%Y') - \
         df_processos['Data de entrada']
```

```python
# tranforma a coluna 'dias da situação' em int
df_processos['dias da situacao'] = df_processos['dias da situacao'].astype(str)
df_processos['dias da situacao'] = df_processos['dias da situacao'].str.replace(' days','', regex=True)
df_processos['dias da situacao'] = df_processos['dias da situacao'].astype('int')
```

```python
# criar uma coluna com o alvo
df_processos.loc[df_processos['Data adesao'].isna(), 'Aderiu'] = False
df_processos.loc[~df_processos['Data adesao'].isna(), 'Aderiu'] = True
df_processos.loc[df_processos['Aderiu']==True]
```

```python
df_processos.drop(columns=['Data de entrada', 'Data adesao'], inplace=True)
```

## Gráfico demonstrando % de adesão por Faixa Etária

```python
qt = (df_processos['Faixa etaria'].value_counts(normalize = True) * 100).\
                to_frame('% por Faixa Etária').reset_index()
qt.rename(columns = {'index': 'Faixa etaria'}, inplace = True)
qt
```

```python
f, ax = plt.subplots()
f.set_size_inches(6, 4)
ax = sns.barplot(y = qt['% por Faixa Etária'], x = qt['Faixa etaria'])
```

```python

gFaixa=df_processos.groupby(['Faixa etaria','Aderiu']).agg({'Processo': 'count'}).\
          reset_index().reset_index().rename(columns={'Processo': 'Total por faixa etária'})

gFaixa=gFaixa.pivot(values='Total por faixa etária',index='Faixa etaria', columns='Aderiu')

gFaixa['% adesão sobre Faixa Etária'] = (gFaixa[True] / (gFaixa[True] + gFaixa[False])*100).\
                                        round(2)

gFaixa['% adesão sobre todas as adesões'] = (gFaixa[True] / gFaixa[True].sum()*100).round(2)

gFaixa.fillna(0, inplace=True)

gFaixa.sort_values('% adesão sobre todas as adesões', ascending=False, inplace=True)

gFaixa
```

```python

#gCarta=df_processos.groupby(['Tipo de Carta','Aderiu']).count()
#gCarta['Total por carta']=df_processos.groupby(['Tipo de Carta']).count()['Nº Processo']
gCarta=df_processos.groupby(['Faixa etaria','Aderiu']).agg({'Processo': 'count'}).reset_index().reset_index().\
                                     rename(columns={'Processo': 'Total por faixa etária'})
gCarta['% resultado por faixa etária'] = 100 * gCarta['Total por faixa etária'] / gCarta.groupby('Faixa etaria')['Total por faixa etária'].transform('sum')

# transformando a linha aderiu em coluna
gCarta=gCarta.reset_index().drop(columns=['index','level_0']).pivot(index='Faixa etaria', columns='Aderiu').rename(columns={'NÃO':'Não aderiu', 'SIM':'Aderiu'}).reset_index()

gCarta.fillna(0, inplace=True)
gCarta
```

```python
x, y = 'Faixa etaria', 'Aderiu'

df1 = df_processos.groupby(x)[y].value_counts(normalize = True)

df1 = df1.mul(100)

df1 = df1.rename('Porcentagem').reset_index()
df1['Aderiu']=df1['Aderiu'].map({False:'Não',True:'Sim'})
g = sns.catplot(x = x, y = 'Porcentagem', hue = y, kind = 'bar', data = df1, aspect = 2.5)
g.set_xticklabels(rotation = 60)
g.ax.set_ylim(0, 100)

for p in g.ax.patches:
    txt = str(p.get_height().round(2)) + '%'
    txt_x = p.get_x() + 0.03
    txt_y = p.get_height() + 1.5
    g.ax.text(txt_x, txt_y, txt, rotation = 90)
```
