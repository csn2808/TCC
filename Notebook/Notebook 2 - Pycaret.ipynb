```python
# estratificando os dados para o treinamento ser mais rápido, mas para apresentação do modelo final, deixar o dataset inteiro

from sklearn.model_selection import StratifiedShuffleSplit

# usa amostragem estratificada para separar os dados em treino e teste
split = StratifiedShuffleSplit(n_splits=1, test_size=0.04, train_size=0.16, random_state=42)
for train_index, test_index in split.split(df_processos, df_processos['Aderiu']):
    strat_train_set = df_processos.loc[train_index]
    strat_test_set = df_processos.loc[test_index]
    
print('train set:', len(strat_train_set))
print('test set: ', len(strat_test_set))
```

```python
df_treino=strat_train_set
df_teste=strat_test_set
```

## Setup do Pycaret

```python
setup_de_classificador = setup(\
                               data=df_processos,\
                               target='Aderiu',\
                               session_id=42,\
                               feature_selection=True,\
                               normalize=True,\
                               #normalize_method = 'minmax', \
                               remove_multicollinearity=True,\
                               remove_perfect_collinearity=True,\
                               remove_outliers = True, \
                               ignore_features=['Processo',\
                                                'Tributo',\
                                                '% multa', \
                                               ], \
                              )
my_setup= pull()
```

```python
# aplicando SMOTE

smt = SMOTE(random_state=42)
Xtrain = get_config('X_train')
ytrain = get_config('y_train')

Xtest = get_config('X_test')
ytest = get_config('y_test')

print('Antes do over sampling ', Xtrain.shape, ytrain.shape)
Xtrain_sampled, ytrain_sampled = smt.fit_sample(Xtrain, ytrain)
Xtest_sampled, ytest_sampled = smt.fit_sample(Xtest,ytest)

set_config('X_train', Xtrain_sampled)
set_config('y_train', ytrain_sampled)
set_config('X_test', Xtest_sampled)
set_config('y_test', ytest_sampled)

Xtrain = get_config('X_train')
ytrain = get_config('y_train')

Xtest = get_config('X_test')
ytest = get_config('y_test')

print('Após over sampling ', Xtrain.shape, ytrain.shape)
```

```python
yi = ytrain.astype(int)
np.bincount(yi)
```

```python
yi2 = ytest.astype(int)
np.bincount(yi2)
```

```python
remove_metric('Kappa')
remove_metric('MCC')
```

```python
models=compare_models(sort = 'F1', # Classificaremos os resultados por esta métrica                      
                      include = ['gbc',     # Gradient Boosting Classifier
                                 'lr',      # Logistic Regression
                                 'ada',     # Ada Boost Classifier
                                 'xgboost', # Extreme Gradient Boosting
                                 'nb',      # Naive Bayes
                                 'rf',      # Random Forest Classifier
                                 'et',      # Extra Trees Classifier
                                 'knn',     # K Neighbors Classifier
                                 'svm',     # SVM - Linear Kernel
                                 ],          
                      n_select = 5        
        )
results=pull()
```

```python
for model in models:
    print(models[2])
```

Após aplicação do SMOTE e com os dados balanceados passaremos ao crossvalidation de todos os modelos gerados no compare_models

```python
tuned_models = {}
results = {}
for model in models:
    tuned_models[str(type(model))] = tune_model(model, \
        optimize = 'F1', n_iter = 200, choose_better = True)
    results[str(type(model))] = pull()
```

```python
for model in models:
    print(tuned_models[str(type(model))])
    display(results[str(type(model))])
    print('==========================================================\n')
```

```python
tuned_results = pd.DataFrame.from_dict({str(tuned_models[key]).split('(')[0]: [value['Accuracy']['Mean'],
                                                                               value['AUC']['Mean'],
                                                                               value['Recall']['Mean'],
                                                                               value['Prec.']['Mean'],
                                                                               value['F1']['Mean']]
                                        for (key, value) in results.items()},
                                        orient=  'index',
                                        columns = ['Accuracy', 'AUC', 'Recall', 'Precision', 'F1'])
```
